## Research Question:
How do career progression factors (years since last promotion, years in current role, ect...) impact the number of years employees stay at the company?

## Hypothesis
### T-Tests:
#### **Two-Sample, One-Sided T-Test: 
- **Null Hypothesis (\( H_0 \))**: The mean `YearsAtCompany` for employees who have left the company is **greater than or equal to** that for employees who have stayed.
- **Alternative Hypothesis (\( H_1 \))**: The mean `YearsAtCompany` for employees who have left the company is **less than** that for employees who have stayed.

#### **Two-Sample, Two-Tailed T-Test: 
- **Null Hypothesis (\( H_0 \))**: The mean `YearsAtCompany` for employees who have left the company is **equal to** the mean for employees who have stayed.
- **Alternative Hypothesis (\( H_1 \))**: The mean `YearsAtCompany` for employees who have left the company is **not equal to** the mean for employees who have stayed.

### Regression Models:
- **Null Hypothesis (\( H_0 \))**: Each predictor variable (Monthly Income, Age, Job Level, Stock Option Level, Years in Current Role, and Years Since Last Promotion) has no statistically significant effect on the number of years an employee stays at the company.
- **Alternative Hypothesis (\( H_1 \))**: At least one predictor variable has a statistically significant effect on the number of years an employee stays at the company.

```python
# Drop Monthly Income as it is too difficult to read
hr_df_dropped = hr_df.drop('MonthlyIncome', axis = 1)
# Function to plot absolute and relative frequency for a given variable
def plot_frequency(variable, title):
    # Calculate absolute frequency
    abs_freq = hr_df[variable].value_counts().sort_index()  # Sort by index
    
    # Calculate relative frequency
    rel_freq = hr_df[variable].value_counts(normalize=True).sort_index() * 100  # Sort by index
    
    # Plotting
    plt.figure(figsize=(14, 6))

    # Absolute frequency
    plt.subplot(1, 2, 1)
    bars_abs = abs_freq.plot(kind='bar', color='skyblue')
    plt.title(f'Absolute Frequency of {title}')
    plt.xlabel(title)
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--')

    # Add values inside the bars for absolute frequency
    for bar in bars_abs.patches:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
                 int(bar.get_height()), ha='center', va='bottom')

    # Relative frequency
    plt.subplot(1, 2, 2)
    bars_rel = rel_freq.plot(kind='bar', color='salmon')
    plt.title(f'Relative Frequency of {title} (%)')
    plt.xlabel(title)
    plt.ylabel('Percentage')
    plt.grid(axis='y', linestyle='--')

    # Add values inside the bars for relative frequency
    for bar in bars_rel.patches:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
                 f'{bar.get_height():.1f}', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

for i in hr_df_dropped:
    plot_frequency(i, i)
```

## 4.3 Two-Sample, One-Tailed T-Test

- **Null Hypothesis (\( H_0 \))**: The mean `YearsAtCompany` for employees who have left the company is **greater than or equal to** that for employees who have stayed.
- **Alternative Hypothesis (\( H_1 \))**: The mean `YearsAtCompany` for employees who have left the company is **less than** that for employees who have stayed.

- **Rationale for the Test**:
  - We selected this test to assess whether the attrition of employees correlates with significantly lower tenure in the company. Understanding this relationship can provide insights into factors affecting employee retention and inform HR strategies. This also indicate whether employees are leaving sooner or later than those who stay.

```python
# Two Sample One Sided T Test to see if YearsAtCompany(Attrition = True) < YearsAtCompany(Attrition = False)

# Identify both samples 
sample_attrition = sampled_users[sampled_users['Attrition']=='Yes']
sample_no_attrition = sampled_users[sampled_users['Attrition']=='No']

# Compute the population and sample mean
population_mean_attrtion = hr_df[hr_df['Attrition']=='Yes']['YearsAtCompany'].mean()
population_mean_NoAttrition = hr_df[hr_df['Attrition']=='No']['YearsAtCompany'].mean()

sample_mean_attrition = sample_attrition.YearsAtCompany.mean()
sample_mean_NoAttrition = sample_no_attrition.YearsAtCompany.mean()

# Compute the one-sided t-test to check if the sample mean differs from the hypothesized mean.
t_statistic_1s_g, p_value_1s_g = stats.ttest_ind(sample_attrition.YearsAtCompany, sample_no_attrition.YearsAtCompany, alternative="greater") 
t_statistic_1s_l, p_value_1s_l = stats.ttest_ind(sample_attrition.YearsAtCompany, sample_no_attrition.YearsAtCompany, alternative="less") 

# Print results
print(f"Population Mean Attrition: {population_mean_attrtion:.2f}")
print(f"Population Mean No Attrition: {population_mean_NoAttrition:.2f}")
print(f"Sample Mean Attrition: {sample_mean_attrition:.2f}")
print(f"Sample Mean No Attrition: {sample_mean_NoAttrition:.2f}")

# One-sided greater
print("\nOne-Sided Test (Attrition Mean > No-Attrition Mean):")
print(f"t-statistic greater: {t_statistic_1s_g:.2f}")
print(f"p-value greater: {p_value_1s_g:.4f}")

# One-sided less
print("\nOne-Sided Test (Attrition Mean < No-Attrition Mean):")
print(f"t-statistic less: {t_statistic_1s_l:.2f}")
print(f"p-value less: {p_value_1s_l:.4f}")
```

- **p-value**: `0.0170`
  - Since `p-value < 0.05`, we **reject the null hypothesis**. This suggests that there is **significant evidence** to support the claim that the mean `YearsAtCompany` for employees who left is **less than** that for employees who stayed.

- **Conclusion**: The mean `YearsAtCompany` for employees who left (5.13 years) is significantly less than that for those who stayed (7.76 years). This finding implies that employees who leave the company tend to have shorter tenures, indicating a potential need for enhanced retention strategies.

## 4.4 Two-Sample, Two-Tailed T-Test

- **Null Hypothesis (\( H_0 \))**: The mean `YearsAtCompany` for employees who have left the company is **equal to** the mean for employees who have stayed.
- **Alternative Hypothesis (\( H_1 \))**: The mean `YearsAtCompany` for employees who have left the company is **not equal to** the mean for employees who have stayed.

- **Rationale for the Test**:
  - We selected this test to evaluate whether the attrition of employees correlates with significant differences in tenure. By comparing these two groups, we can identify the potential impact of attrition on employee retention.

```python
# Step 1: Randomly sample 200 employees from the dataset (ensure you adjust the sample size as needed)
sample_size = 200
random_sample = hr_df.sample(n=sample_size, random_state=42)  # Set random_state for reproducibility

# Step 2: Identify both samples based on attrition status
sample_attrition = random_sample[random_sample['Attrition'] == 'Yes']
sample_no_attrition = random_sample[random_sample['Attrition'] == 'No']

# Step 3: Compute the population and sample means for both groups
population_mean_attrition = hr_df[hr_df['Attrition'] == 'Yes']['YearsAtCompany'].mean()
population_mean_no_attrition = hr_df[hr_df['Attrition'] == 'No']['YearsAtCompany'].mean()

sample_mean_attrition = sample_attrition['YearsAtCompany'].mean()
sample_mean_no_attrition = sample_no_attrition['YearsAtCompany'].mean()

# Step 4: Compute the two-sided t-test
t_statistic_2s, p_value_2s = stats.ttest_ind(sample_attrition['YearsAtCompany'], 
                                              sample_no_attrition['YearsAtCompany'], 
                                              alternative="two-sided")

# Step 5: Print results
print(f"Population Mean Attrition: {population_mean_attrition:.2f}")
print(f"Population Mean No Attrition: {population_mean_no_attrition:.2f}")
print(f"Sample Mean Attrition: {sample_mean_attrition:.2f}")
print(f"Sample Mean No Attrition: {sample_mean_no_attrition:.2f}")

# Two-sided test results
print("\nTwo-Sided Test (Attrition Mean != No-Attrition Mean):")
print(f"t-statistic: {t_statistic_2s:.2f}")
print(f"p-value: {p_value_2s:.4f}")

```

- **p-value**: `0.0340`
  - Since `p-value < 0.05` (alpha level), we **reject the null hypothesis**. This suggests that there is **significant evidence** to support the claim that the mean `YearsAtCompany` for employees who left is **different from** that for employees who stayed.

- **Conclusion**: The results indicate that employees who have left the company (mean tenure of 5.13 years) have significantly shorter tenures compared to those who have stayed (mean tenure of 7.76 years). This finding reinforces the idea that attrition is associated with lower tenure and highlights the need for the organization to investigate the factors leading to employee departures.


In our analysis, we initially performed two one-sample t-tests to evaluate the mean difference of Years at Company within the entire dataset. However, after reviewing the results, we found that these tests did not provide any meaningful insights in the context of our research question, which focuses on comparing employee attrition behavior based on different factors. The one-sample t-tests lacked relevance since they do not allow us to compare differences between groups (i.e., employees who left versus those who stayed), which is central to our analysis.

For this reason, we decided to focus on the two-sample t-tests, which provide a more direct comparison between groups and offer clearer insights into the factors affecting employee attrition. Therefore, the one-sample t-test results were excluded from our final interpretation and report.
