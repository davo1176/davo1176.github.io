```python
import requests
import camelot
import os

# List of PDF URLs to download and extract tables from
pdf_urls = [
    "https://nsaa.org/webdocs/Media_Public/IndustryStats/Historical_Skier_Visits_2024.pdf",
    "https://nsaa.org/webdocs/Media_Public/IndustryStats/ski_areas_per_season_2024.pdf"
     ""    
    # Add more URLs here
]

# Directory to store downloaded PDFs and extracted CSVs
output_dir = "pdf_extraction"
os.makedirs(output_dir, exist_ok=True)

for url in pdf_urls:
    try:
        # Extract the filename from the URL
        pdf_filename = os.path.join(output_dir, url.split("/")[-1])

        # Download the PDF
        response = requests.get(url, stream=True)
        if response.status_code == 200:
            with open(pdf_filename, "wb") as f:
                f.write(response.content)
            print(f"Downloaded: {pdf_filename}")
        else:
            print(f"Failed to download {url}")
            continue

        # Extract tables using Camelot
        tables = camelot.read_pdf(pdf_filename, pages="all", flavor="stream")

        if tables.n > 0:
            for i, table in enumerate(tables):
                csv_filename = f"{pdf_filename.replace('.pdf', '')}_table_{i+1}.csv"
                table.to_csv(csv_filename)
                print(f"Extracted table {i+1} and saved as: {csv_filename}")
        else:
            print(f"No tables found in {pdf_filename}")

    except Exception as e:
        print(f"Error processing {url}: {e}")

print("Processing complete.")
```
